```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import math
import seaborn as sns
```

```python
data=pd.read_csv('./Data/iris.data')
data.columns=['SepalLength','SepalWidth','PetalLength','PetalWidth','Class']
print data.describe()
```

```python
# convert pandas DataFrame to simple numpy arrays
X = data[[0,1,2,3]].values
Y = data['Class'].values
labels = {}
for i,v in enumerate(np.unique(Y)):
    labels[v] = i
    Y[Y==v] = i
Y= Y.astype(np.int8)
```

## Pie chart

Plot the distribution of classes as pie chart. See the documentation for details of pie plot.

```python
plt.pie([sum(Y==labels.values()[0]), sum(Y==labels.values()[1]),sum(Y==labels.values()[2])], labels=labels.keys(), colors=('yellowgreen', 'lightskyblue', 'gold'))
plt.legend()
plt.title('Class distribution of the 3 different flower species')
plt.show()
```

## Bar plot

```python
pos = np.arange(4)
colors = ['b','r','g']
width = 0.2
fig,ax = plt.subplots()

for k,v in labels.iteritems():
    data_class = X[Y==v]
    means = data_class.mean(axis = 0)
    ax.bar(pos + v*width ,means,width=width, label=k, color=colors[v], alpha=0.5)
    ax.set_ylabel('cm')
    ax.set_title('Average values for the flower dimensions')
    ax.set_xticks([p + 1.5 * width for p in pos])
    ax.set_xticklabels(['Sepal length', 'sepal width', 'Petal length','Sepal length' ])
plt.legend()
plt.show()
```

## 1D Histogram

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,6))
counter = 0
titles = ['Sepal length', 'sepal width', 'Petal length','Sepal length' ]
for ax in axes.flatten():
    ax.hist(X[:,counter],25)
    ax.set_xlabel(titles[counter] + " in cm")
    counter += 1

axes[0][0].set_ylabel("Freq")
axes[1][0].set_ylabel("Freq")
fig,tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,6))
counter = 0
colors = ['b','r','g']
titles = ['Sepal length', 'sepal width', 'Petal length','Sepal length' ]
for ax in axes.flatten():
    min_b = math.floor(np.min(X[:,counter]))
    max_b = math.ceil(np.max(X[:,counter]))
    bins = np.linspace(min_b, max_b, 25)
    
    for k,v in labels.iteritems():
        ax.hist(X[Y==v,counter],bins=bins, color = colors[v], alpha = 0.5, label=k)
        ax.set_xlabel(titles[counter] + " in cm")
        ax.legend()
    counter += 1

axes[0][0].set_ylabel("Freq")
axes[1][0].set_ylabel("Freq")

fig,tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,6))
counter = 0
titles = ['Sepal length', 'sepal width', 'Petal length','Sepal length' ]
for ax in axes.flatten():
    
    sns.kdeplot(X[:,counter], ax=ax)
    ax.set_xlabel(titles[counter] + " in cm")
    counter += 1

axes[0][0].set_ylabel("Freq")
axes[1][0].set_ylabel("Freq")
fig,tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,6))
counter = 0
colors = ['b','r','g']
titles = ['Sepal length', 'sepal width', 'Petal length','Sepal length' ]
for ax in axes.flatten():
    
    for k,v in labels.iteritems():
        sns.kdeplot(X[Y==v,counter], color = colors[v],ax=ax, alpha = 0.5, label=k)
        ax.set_xlabel(titles[counter] + " in cm")
        ax.legend()
    counter += 1

axes[0][0].set_ylabel("Freq")
axes[1][0].set_ylabel("Freq")

fig,tight_layout()
plt.show()
```

## Scatter plot

```python
pos = np.arange(4)
colors = ['b','r','g']
markers = ['o','^','s']
width = 0.2
fig,ax = plt.subplots()

for k,v in labels.iteritems():
    ax.scatter(X[Y==v,2], X[Y==v,3], label=k, color=colors[v], alpha=0.5, marker= markers[v])
    ax.set_ylabel('cm')
    ax.set_title('Average values for the flower dimensions')
    ax.set_xticks([p + 1.5 * width for p in pos])
    ax.set_xticklabels(['Sepal length', 'sepal width', 'Petal length','Sepal length' ])
plt.legend()
plt.show()
```

## 3D Scatter plot

```python
from mpl_toolkits.mplot3d import Axes3D

### plot scatterplot

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='3d')
colors = ['blue', 'red', 'green']
markers = ['s', 'o', '^']

for k,v in labels.iteritems():
    ax.scatter(X[Y==v,1],X[Y==v,2], X[Y==v,3], label=k, c=colors[v], alpha=0.5, marker= markers[v])

ax.set_xlabel('sepal width (cm)')
ax.set_ylabel('petal height (cm)')
ax.set_zlabel('petal width (cm)')
plt.show()
```

```python
def boxPlotByClass(data,axes,classlabel):
    #Plots a box plot arranged by class...
    idx=data['Class']==classlabel
    pdata=data[idx][['SepalLength','SepalWidth','PetalWidth','PetalLength']].dropna();
    axes.boxplot(np.asarray(pdata))
# plt.set_xlabel()
    axes.set_xticklabels(['SLength','SWidth','PWidth','PLength'])
    axes.set_title('Class={}'.format(classlabel))
```

### Plot the box-plot class wise as well... i,e.

```python
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(14,6),squeeze=False,sharey=True,)

classes=np.unique(data['Class'].dropna())

boxPlotByClass(data,axes[0,0],classes[0])
boxPlotByClass(data,axes[0,1],classes[1])
boxPlotByClass(data,axes[0,2],classes[2])
```

#### You can also boxplot each feature categorized by the classes...

```python
def boxPlotByFeature(data,axes,xlabel):
    #Plots a box plot for a given feature by class...
    classes=np.unique(data['Class'].dropna())
    legends=[c for c in classes]
    x=data[data['Class']==classes[0]][xlabel].dropna();
    y=data[data['Class']==classes[1]][xlabel].dropna();
    z=data[data['Class']==classes[2]][xlabel].dropna();
    t=[]
    t.append(list(x))
    t.append(list(y))
    t.append(list(z))
    axes.boxplot(t)
    axes.set_xticklabels(legends)
    axes.set_title('Feature={}'.format(xlabel))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(14,12),squeeze=False,sharey=True,)

features=['SepalLength','SepalWidth','PetalWidth','PetalLength']

boxPlotByFeature(data,axes[0,0],features[0])
boxPlotByFeature(data,axes[0,1],features[1])
boxPlotByFeature(data,axes[1,0],features[2])
boxPlotByFeature(data,axes[1,1],features[3])
```
